```python
# Import library
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load dataset
dataset = loadtxt('heart_dataset.csv', delimiter=',')

# Split into input (X) and output (y) variables
X = dataset[:,0:13]
y = dataset[:,13]
# Define the keras model
model = Sequential()

#First hidden layer set to 13 nodes with relu activation function.
model.add(Dense(13,input_shape=(13,),activation='sigmoid') )

#Second hidden layer set to 25 nodes with relu activation function.
model.add(Dense(10, activation='sigmoid'))

#Output layer set to 1 nodes with sigmoid activation function.
model.add(Dense(1, activation='sigmoid'))

# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(X, y, epochs=1000, batch_size=5)

# evaluate the keras model
_,accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python

predictions = (model.predict(X) > 0.5).astype(int)
for i in range(500):
    if predictions[i] == y[i]:
        print('[%d] %s => %d (expected %d) BETUL' % (i,X[i].tolist(), predictions[i], y[i]))
    else:
        print('[%d] %s => %d (expected %d) SALAH' % (i,X[i].tolist(), predictions[i], y[i]))
```
